### 3. Quick look at the data (2-3 real numbers)

Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.

```python
import pandas as pd
import os

# Load the starter CSV
df = pd.read_csv('../../data/raw/content_refresh_anonymized.csv')

# Show 2-3 real numbers that make this lane look worth it
total_content = len(df)
decaying_content = len(df[df['trend_direction'] == 'down'])
avg_decay_pct = df[df['trend_direction'] == 'down']['trend_pct'].mean()

print(f"Total content pieces: {total_content}")
print(f"Pieces with downward traffic trend: {decaying_content} ({decaying_content/total_content*100:.1f}%)")
print(f"Average traffic drop among decaying content: {avg_decay_pct:.1f}%")
```
