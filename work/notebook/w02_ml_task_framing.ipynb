## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

**Scoring / Binary Classification**: We will treat this as a binary classification problem (predicting if a piece of content is 'decaying' or not) and output a probability score. This allows us to score and rank all existing content by their likelihood of decay, helping editors prioritize what to refresh first.

```python
import pandas as pd
import numpy as np

# Checking class balance for classification
df = pd.read_csv('../../data/raw/content_refresh_anonymized.csv')
df['is_decaying'] = (df['trend_direction'] == 'down').astype(int)
print(f"Decaying class balance: {df['is_decaying'].mean()*100:.1f}%")
```

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

**Target variable**: `is_decaying` (Binary 1 or 0).
**Source**: This label comes from a defined rule applied to observed traffic outcomes. Specifically, we define a piece of content as decaying if its traffic in the last 30 days has dropped significantly compared to the previous period (e.g., `trend_direction == 'down'`).

```python
# Example of creating the target label from observed data
df['target_is_decaying'] = np.where((df['trend_direction'] == 'down') & (df['trend_pct'] < -10), 1, 0)
print(f"Target positive rate: {df['target_is_decaying'].mean()*100:.1f}%")
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

**Metric**: Precision at Top K (or Precision on the top 20% highest-scored content).
**Why**: Resources for updating content are limited. We care most about avoiding false positives among the top recommendations sent to writers. If we recommend 100 articles to be refreshed, a precision of >80% means at least 80 of them were genuinely decaying, making it a very efficient use of editorial time.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

An if-statement (e.g., "if age > 365 days and traffic down 20%") ignores the complex interplay of features. For instance, an old piece of content with high competition, low engagement, and short word count might decay much faster than a comprehensive evergreen guide with low competition. ML can learn these non-linear interactions across content attributes, user engagement signals, and competition levels, which is far too messy to hardcode.
